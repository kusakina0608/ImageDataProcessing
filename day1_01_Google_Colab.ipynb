## Alpha Blending

```
import cv2
from google.colab.patches import cv2_imshow

img1 = cv2.imread('/content/drive/My Drive/images/city.jpg', cv2.IMREAD_COLOR)
img2 = cv2.imread('/content/drive/My Drive/images/beach.jpg', cv2.IMREAD_COLOR)
```

```
ch, cw, _ = img1.shape
bh, bw, _ = img2.shape
print('city img shape: ', ch, cw)
print('beach img shape: ', bh, bw)
```

```
add_result = cv2.add(img1, img2[:,:cw])
cv2_imshow(add_result)
```

```
blend_result = img1*0.5 + img2[:,:cw]*0.5
cv2_imshow(blend_result)
```

```
blend_result = img1*0.2 + img2[:,:cw]*0.8
cv2_imshow(blend_result)
```

```
alpha = 0.3
blend_result3 = cv2.addWeighted(img1, alpha, img2[:, :cw], 1-alpha, 0)
cv2_imshow(blend_result3)
```

```
img2[:,:cw] = blend_result3
cv2_imshow(img2)
```

## Bitwise operation

```
import cv2
from google.colab.patches import cv2_imshow

city_img = cv2.imread('/content/drive/My Drive/images/city.jpg', cv2.IMREAD_COLOR)

mask = np.zeros_like(city_img)
cv2.circle(mask, (160,240), 100, (255,255,255), -1)
masked_city = cv2.bitwise_and(city_img, mask)
cv2_imshow(masked_city)
```
